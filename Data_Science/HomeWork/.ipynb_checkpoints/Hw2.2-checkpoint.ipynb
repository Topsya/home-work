```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
tabl = pd.read_csv("2017_jun_final.csv")

# Прочитайте отриману таблицю, використовуючи метод head
tabl.head()
```

```python
# Визначте розмір таблиці за допомогою методу shape
tabl.shape
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
tabl.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)0
tabl_sum_null= tabl.isnull().sum()
tabl_sum_null
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
tabl_del= tabl.drop(labels=["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1)
# tabl_del
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і 
# переконайтеся, що залишився тільки стовпець "Мова.програмування"
new_tabl_del = tabl_del.isnull().sum()
new_tabl_del
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
new_tabl_del2= tabl_del.dropna()
# new_tabl_del2
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
new_tabl_del2.shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову 
# програмування Python
python_data = new_tabl_del2[new_tabl_del2['Язык.программирования'] == 'Python']
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape 
python_data.shape
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
python_data.groupby(by='Должность')
python_data
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за 
# допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
new_paython_data= python_data.groupby("Должность").agg({"Зарплата.в.месяц": ["min", "max"]})
new_paython_data
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(data):
  return data.mean()

new_avg = new_paython_data.apply(fill_avg_salary, axis=1, result_type='expand')
new_avg.name = "avg"
new_paython_data2 = pd.concat([new_paython_data, new_avg], axis=1)
new_paython_data2
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
new_paython_data2['avg'].describe()
```

```python
# Збережіть отриману таблицю в CSV файл
new_paython_data2.to_csv("2017_jun_final_updated.csv")
```
